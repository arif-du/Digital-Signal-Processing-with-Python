# **Computing the DFT of a simple signal**

The Discrete Fourier Transform (DFT) is a mathematical algorithm that transforms a discrete-time sequence of N complex numbers (or samples) into an equivalent-length sequence of complex numbers representing the frequency spectrum of the original sequence. The DFT is widely used in signal processing and analysis, particularly in the fields of audio processing, image processing, and digital communications.

image.png

The DFT converts the time-domain signal into the frequency domain by decomposing the signal into a sum of complex exponential functions with different frequencies. The result of the DFT is a sequence of complex numbers that represent the amplitude and phase of each frequency component in the original signal.

```python
import numpy as np

# Create a simple signal
signal = np.array([1, 2, 3, 4])

# Compute the DFT using numpy.fft.fft
dft = np.fft.fft(signal)

# Print the result
print(dft)
```

The code computes the discrete Fourier transform (DFT) of a simple signal using NumPy's fft function. The signal variable contains an array of four values [1, 2, 3, 4]. The np.fft.fft(signal) function call computes the DFT of the signal and stores the result in the dft variable. Finally, the code prints the DFT.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a signal and compute its DFT
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 10 * t) + 0.5 * np.sin(2 * np.pi * 20 * t)
dft = np.fft.fft(x)

# Compute the IDFT of the DFT
x_reconstructed = np.fft.ifft(dft)

# Plot the original signal and the reconstructed signal
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

axs[0].plot(t, x, color='blue')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Original signal')

axs[1].plot(t, x_reconstructed.real, color='red')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Amplitude')
axs[1].set_title('Reconstructed signal')

plt.tight_layout()
plt.show()
```

In this example, we create a signal x consisting of two sine waves with frequencies 10 Hz and 20 Hz, respectively. We then compute the DFT of the signal using np.fft.fft. Next, we compute the IDFT of the DFT using np.fft.ifft, which gives us the reconstructed signal. Finally, we plot the original signal and the reconstructed signal side by side using matplotlib.pyplot.subplots.

Note that the IDFT of the DFT should be equal to the original signal, up to some numerical precision. To compare the original and reconstructed signals, we plot them on separate subplots using matplotlib.pyplot.subplots.

# **Plotting the power spectrum of a signal**

The power spectrum is the squared magnitude of each frequency component of the Discrete Fourier Transform (DFT) of a signal. It can be calculated using the equation you provided:

`power_spectrum = np.abs(dft)**2`

where dft is the complex-valued vector obtained from the DFT of the signal. The np.abs() function calculates the magnitude of each frequency component, and the **2 operator squares the magnitude to obtain the power spectrum.

The power spectrum represents the distribution of signal power across different frequency components. It can help identify which frequency components are dominant in the signal and how much energy is carried by each component. The power spectrum is often plotted as a function of frequency and is a useful tool for analyzing and understanding signals in many different fields, including communication systems, image processing, and audio signal processing.

This code creates a sinusoidal signal with a frequency of 5 Hz and a duration of 1 second. Then, it computes the Discrete Fourier Transform (DFT) of the signal using numpy.fft.fft. The DFT is a transformation that converts a time-domain signal into its frequency-domain representation. Next, the code computes the power spectrum of the signal by taking the absolute value of the DFT and squaring it. Finally, it plots the power spectrum as a function of frequency using matplotlib.pyplot.plot. The numpy.fft.fftfreq function is used to obtain the frequencies corresponding to each point in the DFT.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a sinusoidal signal
t = np.linspace(0, 1, 100)
signal = np.sin(2 * np.pi * 5 * t)

# Compute the DFT using numpy.fft.fft
dft = np.fft.fft(signal)

# Compute the power spectrum
power_spectrum = np.abs(dft)**2

# Plot the power spectrum
freqs = np.fft.fftfreq(len(signal), t[1] - t[0])
plt.plot(freqs, power_spectrum)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')
plt.show()
```

# **Computing the spectrogram of a signal**

A spectrogram of a signal is a visual representation of the frequencies present in a signal over time. It is obtained by computing the short-time Fourier transform (STFT) of the signal, which involves dividing the signal into short, overlapping windows and computing the Fourier transform of each window. The resulting spectra are then stacked together to form a spectrogram, with time on the x-axis and frequency on the y-axis. The color of each point in the spectrogram represents the magnitude or power of the corresponding frequency component. Spectrograms are commonly used in audio signal processing to analyze the spectral content of sound signals over time.

In this example, we generate a signal with three sinusoids of frequencies 10 Hz, 50 Hz, and 100 Hz. We then compute the spectrogram of the signal using scipy.signal.spectrogram. The fs argument specifies the sampling frequency of the signal, and nperseg and noverlap specify the length of each segment and the overlap between adjacent segments, respectively. Finally, we plot the spectrogram using matplotlib.pyplot.pcolormesh, and use the cmap argument to specify the colormap. The colorbar() function adds a colorbar to the plot.

```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

# Generate a signal with three sinusoids
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 10 * t) + np.sin(2 * np.pi * 50 * t) + np.sin(2 * np.pi * 100 * t)

# Compute the spectrogram
f, t, Sxx = signal.spectrogram(x, fs=1000, nperseg=128, noverlap=64)

# Plot the spectrogram
plt.pcolormesh(t, f, 10*np.log10(Sxx), cmap='viridis')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.colorbar()
plt.show()
```

**Compute the spectrogram**

`f, t, Sxx = signal.spectrogram(x, fs=1000, nperseg=128, noverlap=64)`

 In this line of code, signal.spectrogram is used to compute the spectrogram of the input signal x.

fs=1000 specifies the sampling frequency of the signal in Hz.

nperseg=128 specifies the length of each segment of the signal to be 128 samples long.

noverlap=64 specifies the number of overlapping samples between adjacent segments. In this case, 50% of the samples overlap with each other.

The output of signal.spectrogram is a tuple of three values:

**f:** an array of frequencies in Hz.

**t:** an array of times corresponding to each column of the spectrogram.

**Sxx:** a 2D array containing the spectrogram, with shape (len(f), len(t)). Each element (i, j) of Sxx contains the power spectral density of the signal in the frequency range f[i] to f[i+1] at time t[j] to t[j+1].

**Plot the spectrogram** 

`plt.pcolormesh(t, f, 10*np.log10(Sxx), cmap='viridis')`

This line of code uses matplotlib.pyplot.pcolormesh to create a color mesh plot of the spectrogram.

t is the array of time values computed by signal.spectrogram.

f is the array of frequency values computed by signal.spectrogram.

`10*np.log10(Sxx)` computes the logarithmic power spectral density of the signal. 10*np.log10 scales the data to a decibel (dB) scale, which is often used in signal processing applications to measure signal power.

`cmap='viridis'` specifies the color map to use for the plot. 'viridis' is a popular color map that ranges from yellow to blue.

Together, these arguments create a 2D color mesh plot of the spectrogram, with the x-axis representing time, the y-axis representing frequency, and the color representing the power spectral density of the signal.

# **Compute the circular convolution of two signals using the DFT**

Circular convolution of two signals can be performed using the DFT in the following steps:

i. Take the DFT of both signals using numpy.fft.fft

ii. Multiply the two DFTs pointwise (element-wise multiplication)

iii. Take the inverse DFT of the product using numpy.fft.ifft

The result of the inverse DFT will be the circular convolution of the two signals. Note that the length of the resulting convolution will be the same as the length of the original signals, and the resulting values will be complex numbers. To get the real-valued convolution, you can take the real part of the result.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create two signals
t = np.linspace(0, 1, 1000)
x1 = np.sin(2 * np.pi * 10 * t) + 0.5 * np.sin(2 * np.pi * 20 * t)
x2 = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 15 * t)

# Compute the circular convolution between the signals using the DFT
X1 = np.fft.fft(x1)
X2 = np.fft.fft(x2)
Xconv = np.fft.ifft(X1*X2)

# Plot the circular convolution
fig, ax = plt.subplots()
ax.plot(np.real(Xconv))
ax.set_xlabel('Sample')
ax.set_ylabel('Circular convolution')
ax.set_title('Circular convolution of x1 and x2')
plt.show()
```

In this example, we create two signals x1 and x2 consisting of two sine waves with frequencies 10 Hz and 20 Hz, and 5 Hz and 15 Hz, respectively. We then compute the DFT of each signal using np.fft.fft.

To compute the circular convolution of the signals using the DFT, we multiply the DFT of x1 with the DFT of x2. We then take the inverse DFT of the product using np.fft.ifft to obtain the circular convolution between the signals.

Finally, we plot the circular convolution as a function of the sample index. The length of the circular convolution is equal to the length of the original signals.

Note that if you want to compute the linear convolution of the signals, you need to zero-pad the signals to twice their original length before computing the DFT. This will give you the linear convolution of the signals, which you can then truncate to the original length.

# **Computing the cross-correlation between two signals**

To compute the cross-correlation between two signals using DFT, we can follow these steps:

i. Compute the DFT of the two input signals using numpy.fft.fft.

ii. Take the conjugate of the DFT of the second signal.

iii. Multiply the two DFTs element-wise.

iv. Compute the inverse DFT of the resulting product using numpy.fft.ifft.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create two signals
t = np.linspace(0, 1, 1000)
x1 = np.sin(2 * np.pi * 10 * t) + 0.5 * np.sin(2 * np.pi * 20 * t)
x2 = np.sin(2 * np.pi * 10 * t + np.pi/4) + 0.5 * np.sin(2 * np.pi * 20 * t + np.pi/4)

# Compute the cross-correlation between the signals using the DFT
X1 = np.fft.fft(x1)
X2 = np.fft.fft(x2)
Xcorr = np.fft.ifft(X1*np.conj(X2))

# Plot the cross-correlation
fig, ax = plt.subplots()
ax.plot(np.real(Xcorr))
ax.set_xlabel('Sample')
ax.set_ylabel('Cross-correlation')
ax.set_title('Cross-correlation of x1 and x2')
plt.show()
```
