# **Finite Impulse Response (FIR) filter**
A Finite Impulse Response (FIR) filter is a type of digital filter used in signal processing. It is characterized by a finite-length impulse response, which means that its output is a weighted sum of a finite number of past and present input samples. The coefficients of the FIR filter are typically computed using a windowing technique that helps to shape the frequency response of the filter.

FIR filters have several desirable properties, such as linear phase response and stability, and are widely used in applications such as audio processing, image processing, and telecommunications.

# Filter Types
Low-pass, high-pass, band-pass, and band-stop filters are all types of linear time-invariant filters used in signal processing. Here's a brief description of each filter:

**Low-pass filter:** A low-pass filter allows low-frequency components of a signal to pass through, while attenuating higher frequencies. It's commonly used to remove high-frequency noise from a signal or to smooth out a signal by removing rapid fluctuations.

**High-pass filter:** A high-pass filter allows high-frequency components of a signal to pass through, while attenuating lower frequencies. It's commonly used to remove low-frequency noise from a signal or to emphasize high-frequency features of a signal.

**Band-pass filter:** A band-pass filter allows a specific range of frequencies (the "pass band") to pass through, while attenuating frequencies outside of this range (the "stop band"). It's commonly used to isolate a specific frequency range in a signal or to remove noise that's outside of this range.

**Band-stop filter:** A band-stop filter (also known as a notch filter) attenuates a specific range of frequencies (the "stop band"), while allowing frequencies outside of this range (the "pass band") to pass through. It's commonly used to remove noise that's concentrated in a specific frequency range.

All of these filters can be designed using different methods, including the window method, the Parks-McClellan algorithm, and the least-squares method.

# The steps for designing an FIR filter are as follows:

**Determine the filter specifications:** This involves specifying the filter type (low-pass, high-pass, band-pass, or band-stop), the cutoff frequency, the transition band (the range of frequencies between the pass band and stop band), and any other relevant specifications.

**Choose a filter design method:** There are several methods for designing FIR filters, including windowing, frequency sampling, and least squares. The choice of method depends on the filter specifications and the desired properties of the filter.

**Compute the filter coefficients:** Once the filter design method is chosen, the filter coefficients are computed using a mathematical formula or an algorithm. The coefficients determine the shape of the filter's frequency response.

**Evaluate the filter performance:** The designed filter should be evaluated to ensure that it meets the desired specifications. This involves analyzing the filter's frequency response, phase response, and impulse response, among other metrics.

**Implement the filter:** Once the filter design is complete, it can be implemented in software or hardware to process signals in real-time.

# Window method
The window method is one of the most common methods for designing FIR filters. The basic idea behind the window method is to multiply a truncated impulse response with a window function, which helps to shape the frequency response of the filter. The steps for designing an FIR filter using the window method are as follows:

i. Determine the filter specifications,including the filter type, cutoff frequency, transition band, and number of filter taps.

ii. Choose a window function, such as Hamming, Hann, Blackman, or Kaiser. The choice of window function depends on the desired properties of the filter, such as stopband attenuation, transition width, and passband ripple.

iii. Compute the ideal impulse response of the filter using a mathematical formula or algorithm. This impulse response is usually infinite in length, but is truncated to a finite number of taps to make it realizable.

iv. Multiply the truncated impulse response by the chosen window function to obtain the final filter coefficients. This step is known as windowing.

v. Normalize the filter coefficients to ensure that the filter has unity gain at DC (i.e., zero frequency).

vi. Evaluate the filter performance by analyzing its frequency response, phase response, and other metrics. If the filter does not meet the desired specifications, adjust the window function or the number of taps and repeat the process.

vii. Implement the filter in software or hardware to process signals in real-time.

# Window Function
Each window function used in the FIR filter design has its own advantages and disadvantages. Here are some of the advantages of the commonly used window functions:

**Rectangular window:** The rectangular window is the simplest window function and has a flat response in the passband. It is easy to compute and provides good stopband rejection, but suffers from high sidelobes in the frequency domain.

**Hamming window:** The Hamming window is a widely used window function that provides a trade-off between main lobe width and sidelobe level. It has a moderate main lobe width and relatively low sidelobes, making it suitable for many applications.

**Hann window:** The Hann window is similar to the Hamming window, but has a wider main lobe and lower sidelobes. It provides better stopband rejection than the Hamming window, but at the expense of a wider main lobe.

**Blackman window:** The Blackman window is a tapered window function that provides good stopband attenuation and low sidelobes. It has a wider main lobe than the Hamming or Hann windows, but provides better overall performance in terms of stopband attenuation.

**Kaiser window:** The Kaiser window is a variable-width window that provides control over the main lobe width and sidelobe level. It is often used in applications where precise control over the frequency response is required, such as in digital communication systems. The Kaiser window also allows for the design of filters with higher stopband attenuation than other window functions.

# Low Pass Filter using Hamming window method

In this code, we define the sample rate fs, desired cutoff frequency cutoff, transition width trans_width, and the number of filter taps numtaps. We then use the firwin function to compute the filter coefficients, which are stored in the taps variable. Finally, we use the freqz function to compute the frequency response of the filter, which is plotted using matplotlib.

````python
import numpy as np
from scipy.signal import firwin, freqz
import matplotlib.pyplot as plt

# Filter requirements
fs = 1000 # Sample rate, Hz
cutoff = 100 # Desired cutoff frequency, Hz
trans_width = 100 # Width of transition from pass band to stop band, Hz
numtaps = 101 # Size of the FIR filter

# Compute the filter coefficients using the firwin function
taps = firwin(numtaps, cutoff, window='hamming', pass_zero=True, fs=fs)

# Compute the frequency response of the filter
w, h = freqz(taps, 1, worN=2000)

# Plot the frequency response
fig, ax1 = plt.subplots()
ax1.set_title('Digital filter frequency response')
ax1.plot(0.5*fs*w/np.pi, np.abs(h), 'b')
ax1.set_ylabel('Amplitude [dB]', color='b')
ax1.set_xlabel('Frequency [Hz]')
ax1.set_ylim([-0.05, 1.05])
ax1.set_xlim([0, 0.5*fs])
ax1.grid()

plt.show()
````

# Low Pass Filter using Blackman window method

````python
import numpy as np
from scipy.signal import firwin, freqz
import matplotlib.pyplot as plt

# Filter requirements
fs = 1000 # Sample rate, Hz
cutoff = 100 # Desired cutoff frequency, Hz
trans_width = 100 # Width of transition from pass band to stop band, Hz
numtaps = 101 # Size of the FIR filter

# Compute the filter coefficients using the firwin function
taps = firwin(numtaps, cutoff, window='blackman', pass_zero=True, fs=fs)

# Compute the frequency response of the filter
w, h = freqz(taps, 1, worN=2000)

# Plot the frequency response
fig, ax1 = plt.subplots()
ax1.set_title('Digital filter frequency response')
ax1.plot(0.5*fs*w/np.pi, np.abs(h), 'b')
ax1.set_ylabel('Amplitude [dB]', color='b')
ax1.set_xlabel('Frequency [Hz]')
ax1.set_ylim([-0.05, 1.05])
ax1.set_xlim([0, 0.5*fs])
ax1.grid()

plt.show()
````
