# **Filtering using Convolution**

Convolution is often used to filter signals to remove noise or extract certain features. For example, a low-pass filter can be applied to a signal by convolving the signal with a kernel that has a frequency response that passes low frequencies and blocks high frequencies.

In this example, we first define a low-pass filter kernel by creating an array of ones and dividing it by the filter size. We then generate a random signal of length signal_length. We use scipy.signal.convolve to filter the signal using the filter kernel with mode='same', which returns an output array of the same shape as the input array. Finally, we plot the original signal and the filtered signal using Matplotlib.

````python
import numpy as np
from scipy.signal import convolve

# Define a low-pass filter
filter_size = 10
filter_kernel = np.ones(filter_size) / filter_size

# Generate a signal
signal_length = 100
signal = np.random.rand(signal_length)

# Filter the signal using convolution
filtered_signal = convolve(signal, filter_kernel, mode='same')

# Plot the original signal and the filtered signal
import matplotlib.pyplot as plt
plt.plot(signal, label='Original Signal')
plt.plot(filtered_signal, label='Filtered Signal')
plt.legend()
plt.show()
````


```
# Define a low-pass filter
filter_size = 10
filter_kernel = np.ones(filter_size) / filter_size
```
filter_size is an integer variable representing the size of the filter kernel.

filter_kernel is a NumPy array of length filter_size representing the filter kernel. 

Here, a low-pass filter is defined as a kernel of ones divided by the filter size, which gives equal weights to all the samples within the filter window. This filter is used to smooth out the input signal by averaging the neighboring samples.


```
# Generate a signal
signal_length = 100
signal = np.random.rand(signal_length)
```
Here, we have defined the length of the signal as signal_length and generated a random signal of that length using `np.random.rand()`. This generates an array of signal_length random numbers between 0 and 1.


```
# Filter the signal using convolution
filtered_signal = convolve(signal, filter_kernel, mode='same')
```
This code snippet filters a signal using convolution. It applies a low-pass filter to a random signal of length signal_length. The filter kernel is defined by filter_kernel, which is a vector of length filter_size with all values equal to 1/filter_size. This creates a simple moving average filter.

The convolution is computed using the convolve function from the numpy library, which takes three arguments: the signal to be convolved, the filter kernel, and the mode parameter, which determines how to handle the edges of the signal. 

In this case, we set mode='same', which pads the edges of the signal with zeros to ensure that the filtered output has the same shape as the input signal.

The result is a filtered signal of the same length as the input signal, with each sample computed as the weighted average of the neighboring samples, with more weight given to the samples closer to the current position. This has the effect of smoothing out high-frequency components in the signal, which can help to reduce noise or other unwanted artifacts.


```
# Plot the original signal and the filtered signal
import matplotlib.pyplot as plt
plt.plot(signal, label='Original Signal')
plt.plot(filtered_signal, label='Filtered Signal')

```
The code first imports the matplotlib.pyplot library as plt. Then, it plots the original signal and filtered signal using plt.plot().

The signal is generated using np.random.rand() function which creates an array of random values between 0 and 1 with the specified signal length.

A low-pass filter is defined by creating a kernel with filter_size number of ones and normalizing it by dividing with the filter_size. The filter_kernel is generated by dividing ones with filter_size.

The convolution of the signal and filter is computed using convolve() function from scipy.signal library with mode='same' which returns the output of the convolution with the same size as the input signal.

Finally, plt.legend() displays the legend on the plot and plt.show() displays the plot on the screen.
