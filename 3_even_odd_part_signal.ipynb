# **Even and Odd part of a Signal**

Even signal definition: When a signal exhibits symmetry with respect to t=0, then it is called an
even signal. The even signal will satisfy the condition, x(−n) = x(n).

image.png

Odd signal definition: When a signal exhibits antisymmetry with respect to t=0, then it is called an
odd signal. The odd signal will satisfy the condition, x(−n) = −x(n)

image.png

Folded Signal definition: The signal x(n) that flips around n = 0 is called folded signal which is
denoted by x(−n).

# **Check whether a given signal is even or odd**
To check whether a given signal is even or odd, you can apply the following two tests:

**Even signal test:** A signal x(t) is even if it satisfies the condition `x(-t) = x(t)` for all t.

**Odd signal test:** A signal x(t) is odd if it satisfies the condition `x(-t) = -x(t)` for all t.

To apply these tests in practice, you can evaluate the signal at negative values of t, and compare it to the corresponding values at positive values of t. 

If the signal values are equal at negative and positive t, the signal is even. 

If the signal values are equal in magnitude but opposite in sign at negative and positive t, the signal is odd.

Here's an example Python code that checks whether a signal is even or odd:

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal
t = np.linspace(-5, 5, 1000)
x = np.cos(t)

# Evaluate the signal at positive and negative t
x_neg = x[::-1]   # reverse the signal
t_neg = -t        # negate the time axis
x_pos = x
t_pos = t

# Check if the signal is even or odd
if np.allclose(x_pos, x_neg):
    print("Signal is even")
elif np.allclose(x_pos, -x_neg):
    print("Signal is odd")
else:
    print("Signal is neither even nor odd")

# Plot the signal and its symmetry properties
plt.plot(t, x, label='x(t)')
plt.plot(t_neg, x_neg, label='x(-t)')
plt.plot(t_pos, x_pos, label='x(t)')
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal Symmetry Test')
plt.grid()
plt.show()
````

In this example, the signal x is defined as a cosine wave. The signal is then evaluated at positive and negative values of t using array indexing and negation of the time axis. The `np.allclose()` function is used to compare the signal values at positive and negative t, and determine whether the signal is even or odd. Finally, the signal and its symmetry properties are plotted using Matplotlib.

# **Even and Odd Part**

````python
import numpy as np
import matplotlib.pyplot as plt

def even_odd(x):
    x_e = 0.5 * (x + np.flip(x))
    x_o = 0.5 * (x - np.flip(x))
    return x_e, x_o

t = np.linspace(-5, 5, 1000)
x = np.sin(2*np.pi*t) + np.cos(4*np.pi*t)

x_e, x_o = even_odd(x)

plt.plot(t, x, label='Original signal')
plt.plot(t, x_e, label='Even part')
plt.plot(t, x_o, label='Odd part')
plt.legend()
plt.show()
````
