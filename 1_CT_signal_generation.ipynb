# **CT Sine Wave Generation**

A **continuous-time (CT)** sine wave is a type of signal that is periodic and has a sinusoidal shape. The general form of a CT sine wave is given by:

`x(t) = A sin(2*pi*f*t + phi)`

where A is the amplitude, f is the frequency, t is time, and phi is the phase angle. The sine wave has a period of T = 1/f and repeats itself after every T seconds.

In Python, you can generate a CT sine wave using the numpy library, which provides the `sin()` function for computing the sine of a given input. 

For example, to generate a CT sine wave with A = 1, f = 2, and phi = 0 in the range [0, 1], you can use the following code:

````python
import numpy as np
import matplotlib.pyplot as plt

A = 1
f = 2
phi = 0
t = np.linspace(0, 1, 100)
x = A * np.sin(2 * np.pi * f * t + phi)

plt.plot(t, x)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.title('Continuous-Time Sine Wave')
plt.show()
````

This code first imports **numpy** and **matplotlib.pyplot** libraries.

It then sets the values of A, f, and phi to 1, 2, and 0, respectively, and generates a time vector t using the `linspace()` function to generate 100 equally spaced time values between 0 and 1. 

The sine wave x is then computed using the `sin()` function, and plotted using `plot()` function of `matplotlib.pyplot`. 

The resulting plot shows the continuous-time sine wave with a frequency of 2 Hz and an amplitude of 1. The phase angle phi determines the position of the wave at t=0. If phi=0, the wave starts at its maximum value, whereas if phi=pi/2, the wave starts at 0.

# **CT Cosine Signal Generation**
A continuous-time (CT) cosine signal is another periodic signal that has a shape similar to a CT sine wave but *with a phase shift of 90 degrees*. The general form of a CT cosine signal is given by:

`x(t) = A cos(2*pi*f*t + phi)`

where A is the amplitude, f is the frequency, t is time, and phi is the phase angle. The cosine wave also has a period of T = 1/f and repeats itself after every T seconds.

In Python, you can generate a CT cosine wave using the numpy library, which provides the **cos()** function for computing the cosine of a given input. For example, to generate a CT cosine wave with A = 1, f = 2, and phi = 0 in the `range [0, 1]`, you can use the following code:

````python
import numpy as np
import matplotlib.pyplot as plt

A = 1
f = 2
phi = 0
t = np.linspace(0, 1, 100)
x = A * np.cos(2 * np.pi * f * t + phi)

plt.plot(t, x)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.title('Continuous-Time Cosine Wave')
plt.show()
````

This code is similar to the previous example that generated a CT sine wave, but it uses the `cos()` function instead of sin(). 

The resulting plot shows the continuous-time cosine wave with a frequency of 2 Hz and an amplitude of 1. 

The phase angle phi determines the position of the wave at t=0. If phi=0, the wave starts at its maximum value, whereas if phi=pi/2, the wave starts at 0.

# **Deterministic CT signal**
A deterministic continuous-time (CT) signal is a mathematical function that can be expressed using a formula or an equation. 

Such signals have a known and predictable behavior and can be characterized by their amplitude, frequency, and phase. 

The most common examples of deterministic CT signals include the sinusoidal waves (sine and cosine waves), the unit step function, the impulse function, and the ramp function.

To generate a deterministic CT signal in Python, you need to define the formula or equation that describes the signal and then plot the signal using the appropriate plotting function. 

For example, to generate a CT sinusoidal wave with A=1, f=1, phi=0, and a sampling rate of 1000 Hz in the range [0, 1], you can use the following code:

````python
import numpy as np
import matplotlib.pyplot as plt

A = 1
f = 1
phi = 0
fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)
x = A * np.sin(2 * np.pi * f * t + phi)

plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Deterministic CT Sinusoidal Wave')
plt.show()
````

This code generates a CT sinusoidal wave with a frequency of 1 Hz, an amplitude of 1, and a phase of 0. The `np.linspace()` function is used to generate a time vector t that spans from 0 to 1 second with a sampling rate of 1000 Hz. The `np.sin()` function computes the sine of the input using radians, and the resulting signal is plotted using the `plt.plot()` function.

Similarly, you can generate other deterministic CT signals such as the cosine wave, the unit step function, the impulse function, and the ramp function using the appropriate formulas and plotting functions.

# **Random CT Signal**
A random continuous-time (CT) signal is a signal that varies unpredictably over time and has no mathematical formula or equation that describes its behavior. Random signals can be generated using a variety of techniques, including sampling from probability distributions, noise generators, and stochastic processes.

One common type of random CT signal is **white noise**, which is a signal that has a flat power spectral density and is characterized by a constant amplitude across all frequencies. In Python, you can generate white noise using the numpy library, which provides several functions for generating random numbers.

For example, to generate a random CT signal that consists of white noise with a sampling rate of 1000 Hz and a duration of 1 second, you can use the following code:

````python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)
x = np.random.normal(0, 1, len(t))

plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Random CT Signal (White Noise)')
plt.show()
````

# **Periodic Signal**
A periodic signal is a signal that repeats itself exactly after a fixed time interval, called the period. Mathematically, a continuous-time (CT) signal x(t) is periodic with period T if and only if:

`x(t) = x(t + kT)`

for all t and some integer k. This means that the signal has the same values and shape over each period, and the period is the smallest time interval for which this property holds.

Periodic signals are commonly encountered in signal processing and communication systems, as they can be used to represent many types of waveforms, such as sine waves, square waves, and sawtooth waves. They are also important in the context of Fourier analysis, as they can be decomposed into a series of harmonic components using the Fourier series.

In Python, you can generate and visualize periodic signals using the numpy library, which provides many functions for creating and manipulating arrays of numerical data. 

Here are some examples of how to generate and plot periodic signals in Python:

**Sine Wave**

````python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000 # sampling frequency
T = 1/fs # sampling period
f = 10 # frequency of sine wave
t = np.arange(0, 1, T) # time vector

x = np.sin(2*np.pi*f*t) # generate sine wave

plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine wave')
plt.show()
````

This code generates a sine wave with a frequency of 10 Hz and a duration of 1 second, sampled at a rate of 1000 Hz. The sine wave is created using the **np.sin()** function, which takes an array of time values and returns an array of corresponding sine values. The resulting waveform is plotted using the matplotlib library, which provides functions for creating plots and visualizations.

**Square Wave**

````python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000 # sampling frequency
T = 1/fs # sampling period
f = 5 # frequency of square wave
t = np.arange(0, 1, T) # time vector

x = np.sign(np.sin(2*np.pi*f*t)) # generate square wave

plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Square wave')
plt.ylim([-1.2, 1.2])
plt.show()
````

This code generates a square wave with a frequency of 5 Hz and a duration of 1 second, sampled at a rate of 1000 Hz. The square wave is created using the `np.sign()` function, which returns an array of sign values (+1 or -1) based on the sign of the input values. The input values are generated using a sine wave with the same frequency as the square wave. The resulting waveform is plotted using matplotlib, and the y-axis limits are set to [-1.2, 1.2] to ensure that the plot shows the full range of the square wave.

Note that the `ylim()` function is used to set the y-axis limits to [-1.2, 1.2] to ensure that the full amplitude range of the square wave is visible in the plot.

**Sawtooth wave**

````python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt

# Define the sampling rate and time vector
fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

# Generate the sawtooth wave
f = 10 # Frequency in Hz
x = sig.sawtooth(2*np.pi*f*t)

# Plot the waveform
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sawtooth Wave ({} Hz)'.format(f))
plt.show()
````

In this code, the `linspace()` function is used to create a time vector t that spans from 0 to 1 second with a step size of 1/fs seconds. The `sawtooth()` function is then used to generate the sawtooth waveform with a frequency of 10 Hz and a period of 1/10 seconds. The resulting waveform is plotted using `matplotlib.pyplot`. The resulting plot shows the sawtooth wave over the time interval [0, 1] seconds.

# **Aperiodic Signal**
An aperiodic signal is a signal that does not repeat itself at regular intervals. It can be a signal that is finite in duration, or a signal that is infinite in duration but has no repeating pattern. 

Examples of aperiodic signals include a single pulse, a decaying exponential, or a Gaussian function.

Here is an example Python code that generates an aperiodic signal in the form of a Gaussian function:

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the time vector
t = np.linspace(-5, 5, 1000)

# Define the parameters of the Gaussian function
A = 1 # Amplitude
mu = 0 # Mean
sigma = 1 # Standard deviation

# Generate the Gaussian function
x = A * np.exp(-(t-mu)**2 / (2*sigma**2))

# Plot the waveform
plt.plot(t, x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Gaussian Function')
plt.show()
````

In this code, the `linspace()` function is used to create a time vector t that spans from `-5 to 5` with a step size of `(5-(-5))/1000`. The parameters A, mu, and sigma are used to define the Gaussian function, which is generated using the `np.exp()` function. The resulting waveform is plotted using matplotlib.pyplot. The resulting plot shows the aperiodic Gaussian function over the time interval `[-5, 5]`.

# **Symmetric Signal**
A symmetric signal is a signal that is identical to its mirror image about a vertical axis. In other words, if the signal is folded about the vertical axis, the two halves of the signal will be identical. Mathematically, a signal is symmetric if:

`x[n] = x[-n]`

Here is an example Python code that generates a symmetric signal in the form of a triangle wave:

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the time vector
t = np.linspace(-1, 1, 1000)

# Define the period of the triangle wave
T = 0.5

# Generate the triangle wave
x = np.abs((t/T) - 2*np.floor((t/T) + 0.5))

# Generate the symmetric triangle wave
x_sym = 0.5 * (x + np.flip(x))

# Plot the original and symmetric waveforms
plt.plot(t, x, label='Original')
plt.plot(t, x_sym, label='Symmetric')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Symmetric Triangle Wave')
plt.legend()
plt.show()
````

The x variable generates the original triangle wave using the formula:

`x[n] = |(n/T) - 2floor((n/T) + 0.5)|`

where T is the period of the triangle wave.

The x_sym variable generates the symmetric triangle wave by taking the average of x and its mirror image using the formula:

`x_sym[n] = 0.5(x[n] + x[-n])`

where x[-n] is the mirror image of x[n].

By using this formula, we obtain a signal that is identical to its mirror image about a vertical axis.


# **Anti-symmetric Signal**
An antisymmetric signal is a signal that is equal to the negative of its mirror image about a vertical axis. In other words, if the signal is folded about the vertical axis, the two halves of the signal will be mirror images of each other with opposite sign. Mathematically, a signal is antisymmetric if:

`x[n] = -x[-n]`

Here is an example Python code that generates an antisymmetric signal in the form of a square wave:

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the time vector
t = np.linspace(-1, 1, 1000)

# Define the period of the square wave
T = 0.5

# Generate the square wave
x = np.sign(np.sin(2*np.pi*t/T))

# Generate the antisymmetric square wave
x_anti = -0.5 * (x - np.flip(x))

# Plot the original and antisymmetric waveforms
plt.plot(t, x, label='Original')
plt.plot(t, x_anti, label='Antisymmetric')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Antisymmetric Square Wave')
plt.legend()
plt.show()
````

The resulting square wave is then shifted down by 0.5 so that it is centered around 0, and then multiplied by -1 to obtain the antisymmetric square wave. The `np.flip()` function returns a reversed copy of the array, which is then subtracted from x to produce the mirror image with opposite sign. 

Finally, the result is scaled by 0.5 to get the average of the original and mirrored signal.

# **Parabolic Signal**
A parabolic signal is a continuous-time signal that has a quadratic form. It can be expressed mathematically as:

`x(t) = At^2 + Bt + C`

where A, B, and C are constants that determine the shape and position of the parabolic signal.

Here's an example Python code to generate a parabolic signal:

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the constants
A = 1
B = 2
C = 3

# Generate the time axis
t = np.linspace(-1, 1, 1000)

# Generate the parabolic signal
x = A*t**2 + B*t + C

# Plot the signal
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Parabolic Signal')
plt.grid()
plt.show()
````

In this example, the constants A, B, and C are set to 1, 2, and 3 respectively. The time axis t is generated using the `np.linspace()` function, which generates an array of 1000 equally spaced points between -1 and 1. The parabolic signal x is then generated using the quadratic formula, and plotted using the `plt.plot()` function from the matplotlib library. The resulting plot shows a parabolic signal that opens upwards, with its vertex located at (-B/2A, C-B^2/4A).

# **Even Signal**
An even signal is a signal that is symmetric about the y-axis. Mathematically, a signal x(t) is even if it satisfies the condition x`(-t) = x(t)` for all t.

One common example of an even signal is the cosine function, which is even because `cos(-t) = cos(t)` for all t.

Here's an example Python code to generate an even signal:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate the time axis
t = np.linspace(-np.pi, np.pi, 1000)

# Generate the even signal
x = np.cos(t)

# Plot the signal
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Even Signal')
plt.grid()
plt.show()
````

In this example, the time axis t is generated using the `np.linspace()` function, which generates an array of 1000 equally spaced points between -π and π. The even signal x is then generated using the cosine function, which is an even function. The resulting plot shows a cosine wave that is symmetric about the y-axis.

# **Odd Signal**
An odd signal is a signal that is anti-symmetric about the y-axis. Mathematically, a signal x(t) is odd if it satisfies the condition `x(-t) = -x(t)` for all t.

One common example of an odd signal is the sine function, which is odd because `sin(-t) = -sin(t)` for all t.

Here's an example Python code to generate an odd signal:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate the time axis
t = np.linspace(-np.pi, np.pi, 1000)

# Generate the odd signal
x = np.sin(t)

# Plot the signal
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Odd Signal')
plt.grid()
plt.show()
````

In this example, the time axis t is generated using the `np.linspace()` function, which generates an array of 1000 equally spaced points between -π and π. The odd signal x is then generated using the sine function, which is an odd function. The resulting plot shows a sine wave that is anti-symmetric about the y-axis.
